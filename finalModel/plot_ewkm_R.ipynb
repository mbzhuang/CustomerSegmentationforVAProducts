```r
conv_df_boxplot <- function(df){
    df = as.data.frame(t(df))
    names(df) = c("y0", "y25", "y50", "y75", "y100")
    df$cluster_id = paste0("Cluster", " ", c(1:7))
    return(df)
}
```

```r
df_CEN_bg_pctHHincomeLT15K = conv_df_boxplot(df_CEN_bg_pctHHincomeLT15K)
df_CEN_tr_pctHSGrad = conv_df_boxplot(df_CEN_tr_pctHSGrad)
df_EstMarketValue_C1 = conv_df_boxplot(df_EstMarketValue_C1)
df_HealthScore_C5 = conv_df_boxplot(df_HealthScore_C5)
df_score_c4 = conv_df_boxplot(df_score_c4)
```

```r
ggplot(df_CEN_bg_pctHHincomeLT15K, aes(x = cluster_id, fill = cluster_id)) +
  geom_boxplot(
   aes(ymin = y0, lower = y25, middle = y50, upper = y75, ymax = y100),
   stat = "identity"
 ) +
 theme_classic() +
 labs(title = "Percentage of Households Annual Income Less than $15,000 \n",
      x = "",
      y = "Percentage") +
scale_y_continuous(labels = scales::percent) +
    theme(legend.position="none",
          line = element_blank(),
          text=element_text(size=13))
```
